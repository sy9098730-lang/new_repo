```python
# Find common elements between two lists.
li=[2,2,3,3,4,5,6]
li2=[2,3,8,4,9]
li3=[]
for i in li:
    for j in li2:
        if i==j and i not in li3:
            li3.append(i)

          

print(li3)
```

```python
# Rotate a list by k positions.
li=[1,2,3,4,5]
k=2
k=k%len(li)
result=li[-k:] +li[:-k]
print(result)
```
