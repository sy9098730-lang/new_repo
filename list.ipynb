```python
list=[2,9,3,4,5]
```

```python
# Add an element to the end of a list.
list.append(6)
list
```

```python
# Insert an element at the beginning of a list.
list[0]=9
list
```

```python
# Remove an element from a list using remove().
list.remove(9)
list
```

```python
# Print the length of a list.
print(len(list))
```

```python
# Access the first and last element of a list.
first=list[0]
last=list[len(list)-1]
print(first)
print(last)
```

```python
# Check if an element exists in a list.
for i in list:
   if i in list:
        print("yes")
        break
else:
        print("No")
```

```python
# Reverse a list
print(list[-1::-1])
```

```python
# Sort a list in ascending order.
print(list.sort())
list
```

```python
print(list.count(3))
```

```python
# Count how many times an element appears in a list.
dic={}
for i in list:
    if i in dic:
        dic[i]+=1

    else :
        dic[i]=1

print(dic)
```

```python
# Find the largest number in a list (without using max()).
list=[2,4,5,7,1]
min=list[0]
for i in list:
    if min>i:
        min=i

print(min)
```

```python
# Find the sum of all elements in a list.
sum=0
for i in list:
    sum+=i

print(sum)
```
