```python
# ○ Star patterns 
for i in range(5):
    for j in range(5-i):
        print("*",end=" ")
    print()
```

```python
#  ○ Number patterns 
for i in range(4,0,-1):
    for j in range(1,i+i):
        print(j,end="")

    print()
    
```

```python
#  FizzBuzz 
number=input("Enter your number")
if number%3==0:
    print("fizz")
elif number%5==0:
    print("buzz")
elif number%3==0 and number%5==0:
    print("fizzbuzz")
```
